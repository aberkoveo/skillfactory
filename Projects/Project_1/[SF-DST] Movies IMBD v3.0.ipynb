# 34. Какие фильмы входят в 1 процент лучших по рейтингу?
Варианты ответа:
1. Inside Out, Gone Girl, 12 Years a Slave
2. BloodRayne, The Adventures of Rocky & Bullwinkle
3. The Lord of the Rings: The Return of the King
4. 300, Lucky Number Slevin

```python
answers = {
    1: set(['Inside Out', 'Gone Girl', '12 Years a Slave']),
    2: set(['BloodRayne', 'The Adventures of Rocky & Bullwinkle']),
    3: set(['The Lord of the Rings: The Return of the King','Upside Down']), # в оригинальном ноутбуке данный ответ состоит из 'The Lord of the Rings: The Return of the King'
                                                                             # вариант исправлен в соотв. с платформой                                                   
    4: set(['300', 'Lucky Number Slevin'])
}

top_procent_check = 1
top_procent_vote = np.percentile(data['vote_average'], (100-top_procent_check))
top_procent_films = set(data[data.vote_average>top_procent_vote]['original_title'].to_list())
display(data[data.vote_average>top_procent_vote].sort_values(by='vote_average')[['original_title', 'vote_average', 'vote_count']])

for answer in answers:
    if answers[answer].issubset(top_procent_films):
        print(f'Ответ:  {answer}. {answers[answer]}')
    
answer_ls.append(1)
```

# 36. У какого из режиссеров выше вероятность выпустить фильм в прибыли? (5 баллов)101
Варианты ответа:
1. Quentin Tarantino
2. Steven Soderbergh
3. Robert Rodriguez
4. Christopher Nolan
5. Clint Eastwood

```python
answers = {
    1: 'Quentin Tarantino',
    2: 'Steven Soderbergh',
    3: 'Robert Rodriguez',
    4: 'Christopher Nolan',
    5: 'Clint Eastwood'
    }

data_36 = data.copy()
data_36['is_profit'] = data_36.profit.apply(lambda x: 1 if x>0 else 0)
expanded_data = expand_df(data_36, 'director')
director_chances = expanded_data.groupby('director_divided')['is_profit'].agg(
    [('is_profit', sum),
    ('at_all', len),
    ('chance', lambda x: x.sum() / x.count() ) ]
    )
display(director_chances.sort_values(by=['chance','at_all'], ascending=False).head(7))
top_chance_directors_by_order = director_chances.sort_values(by=['chance','at_all'], ascending=False).index.to_list()

for director in top_chance_directors_by_order:
    if director in answers.values():
        print(f'по вероятности на успех {director} на {top_chance_directors_by_order.index(director) + 1} месте')
        print(f'Ответ под номером {list(answers.keys())[list(answers.values()).index(director)]}')
        break

answer_ls.append(4)
```
